```python
import pandas as pd
```

```python
df = pd.read_csv("./240930_Data_source_cas_pratique_stage_data_analyst - Sheet1.csv")
df.head()
```

```python
# Split the 'Transmission Geolocation' column by ',' and extract the latitude
df['latitude'] = df['Transmission Geolocation'].apply(lambda x: float(x.split(',')[0]))
df['longitude'] = df['Transmission Geolocation'].apply(lambda x: float(x.split(',')[1]))
df.drop(columns=['Transmission Geolocation'], inplace=True)
df.rename(columns={df.columns[5]: 'Transmission Accuracy'}, inplace=True)
df.head()
```

```python
df.to_csv("clean.csv", index=False)
```
